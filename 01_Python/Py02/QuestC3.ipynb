```python
# 클로저를 활용하여 함수 호출 횟수를 출력하는 데코레이터를 만들어봅시다.

#데코레이터는 함수를 꾸며주는 함수

def counter(func):  # func 자리에 say_hello 함수가 들어올 예정
    count = 0  # 실행 횟수를 기억할 클로저 변수
    
    # func의 매개변수를 유연하게 받기 위해 *args, **kwargs 사용
    def wrapper(*args, **kwargs):
        nonlocal count 
        
        result = func(*args, **kwargs)
        
        # 실행 횟수 증가 및 출력
        count += 1
        print(f"{func.__name__} 실행횟수: {count}")
        
        return result
        
    return wrapper

# --- 데코레이터 적용 및 테스트 ---
@counter
def say_hello():
    print("Hello Aiffel!")

for i in range(5):
    say_hello()
```
